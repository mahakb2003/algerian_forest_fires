## Algerian Forest Fires Dataset
Data Set Info:
The Dataset include 244 instances that regroup the data of two regions of algeria ,namely the bejaia region located in the northeast of algeria and the Sidi Bel-abbes region located in the northwest of algeria.

 122 instances for each region.
 the period from june 2012 to september 2012 .The Dataset includes 11 attributes and 1 ouput attribute (class).\
 The 244 instances have been classified into 
 138 classes into fire
 106 classes into not fire
 



```python
import pandas as pd
import numpy as numpy
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline
```

## Data Cleaning

```python
dataset.isnull().sum()
```

```python
dataset[dataset.isnull().any(axis=1)]
```

```python
dataset
```

The data is converted into two sets besed on region from 122 ith index,we can make a new column based on region:

1.Bejaia Region Dataset

2.Sidi-Bel Abbes Region Dataset 

Add New Column with Region

```python
dataset.loc[:122,"Region"]=0
dataset.loc[122:,"Region"]=1
dataset.info()
```

```python

dataset[['Region']] = dataset[['Region']].astype(int)
dataset
```

```python
df=dataset.copy()
```

```python
df.isnull().sum()
```

```python
#Remove the null values
df=df.dropna().reset_index(drop=True)
df
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df=df.drop(122).reset_index(drop=True)
df.iloc[[122]]
```

```python
df.columns
```

```python
## Fix Spaces in columns names
df.columns=df.columns.str.strip()
df.columns
```

```python
df.info()
```

```python
df.columns
```

```python
## Change the Require columns as integer datatype.
df[['month','day','year','Temperature','RH','Ws']]=df[['month','day','year','Temperature','RH','Ws']].astype(int)
df.info()
```

```python
df.columns
```

```python
# changing the other columns to float datatype
objects=[features for features in df.columns if df[features].dtype=='O']
for i in objects:
    if i!='Classes':
        df[i]=df[i].astype(float)
```

```python
df.info()
```

```python
df.head()
```

```python
# Check for Duplicate Rows
duplicate_rows = df[df.duplicated()]
duplicate_rows 
```

```python
# let save the cleaned data set to new csv
df.to_csv('Algerian_forest_fires_cleaned_dataset.csv',index=False)
```
